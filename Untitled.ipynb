```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('netflix_titles.csv')

df = df.dropna(subset=['type','release_year','rating','country','duration'])

type_counts = df['type'].value_counts()
plt.figure(figsize=(6,4))

plt.bar(type_counts.index,type_counts.values,color=['skyblue','orange'])

plt.title('numbar is movi vs tv show on netflix')
plt.xlabel('type')

plt.ylabel('count')

plt.tight_layout()

plt.savefig('movi_vs_tv.png')
plt.show()


rating_counts = df['rating'].value_counts()
plt.figure(figsize=(8,6))


plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=90)

plt.title('percentage of ontent rating')


plt.tight_layout()
plt.savefig('percent_rating.png')

plt.show()



move_df = df[df['type'] == 'Movie'].copy()

move_df['duration_int'] = move_df['duration'].str.replace('min','').astype(int)
plt.figure(figsize=(8,6))

plt.hist(move_df['duration_int'],bins=30,color='purple',edgecolor ='black')

plt.title('Distribusan of move duration')


plt.xlabel('duration (minutes)')
plt.ylabel('numbar of movie')

plt.tight_layout()

plt.savefig('durastion_move.png')
plt.show()


release_counts = df ['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.scatter(release_counts.index,release_counts.values,color='red')

plt.title('release count scatter')

plt.xlabel('release count year')
plt.ylabel('numbar of year')

plt.tight_layout()

plt.savefig('release_year.png')
plt.show()


contry_count = df['country'].value_counts().head()
plt.figure(figsize=(8,6))

plt.barh(contry_count.index,contry_count.values,color='red')


plt.title('top 10 country  by numbar of shows')

plt.xlabel('numbar of showe')
plt.ylabel('countery')

plt.tight_layout()

plt.savefig('counter_showe.png')
plt.show()


content_by_year = df.groupby(['release_year','type']).size().unstack().fillna(0)
fig , ax = plt.subplots(1,2,figsize=(12,5))


ax[0].plot(content_by_year.index,content_by_year['Movie'],color='green')
ax[0].set_title('Movies releshd per year')


ax[0].set_xlabel('year')
ax[0].set_ylabel('numbar of movies')


ax[0].plot(content_by_year.index,content_by_year['TV Show'],color='orange')

ax[0].set_title('TV show releshd per year')


ax[0].set_xlabel('year')
ax[0].set_ylabel('numbar of movies')


fig.suptitle('comprison of movies and tv show released over year')
plt.tight_layout()
plt.savefig('tv_show.png')


plt.show()
```
